```python
A = torch.randn(3,3)
print(A)
print(torch.abs(A)) # Ï†àÎåìÍ∞í |x|
print(torch.sqrt(torch.abs(A))) # Ï†úÍ≥±Í∑º(Î£®Ìä∏) \sqrt{x}
print(torch.exp(A)) # ÏûêÏó∞ÏÉÅÏàò ÏßÄÏàò e^x
print(torch.log(torch.abs(A))) # ÏûêÏó∞Î°úÍ∑∏ ln(x)
print(torch.log(torch.exp(torch.tensor(1)))) # ln(e^1) = 1 ÌôïÏù∏
print(torch.log10(torch.tensor(10))) # ÏÉÅÏö©Î°úÍ∑∏ log_10(x)
print(torch.log2(torch.tensor(2))) # Ïù¥ÏßÑÎ°úÍ∑∏ log_2(x)
print(torch.round(A)) # Î∞òÏò¨Î¶º (Ï†ïÏàò)
print(torch.round(A,decimals=2)) # ÏÜåÏàòÏ†ê ÎëòÏß∏ ÏûêÎ¶¨ÍπåÏßÄ Î∞òÏò¨Î¶º
print(torch.floor(A)) # ÎÇ¥Î¶º (Î∞îÎã• Ìï®Ïàò)
print(torch.ceil(A)) # Ïò¨Î¶º (Ï≤úÏû• Ìï®Ïàò)
```

```python
A = torch.randn(3,4)
print(A)
print(torch.max(A))
# üí° ÌïµÏã¨ ÏõêÏπô: "dimÏùÄ ÏßÑÌñâ Î∞©Ìñ•Ïù¥Îã§!"
print(torch.max(A,dim=0)) # Í∞Å 'Ïó¥'Ïùò ÏµúÎåìÍ∞í (ÏúÑÏïÑÎûòÎ°ú Íæπ ÎàÑÎ•¥Í∏∞)
print(torch.max(A,dim=1)) # Í∞Å 'Ìñâ'Ïùò ÏµúÎåìÍ∞í (ÏñëÏòÜÏóêÏÑú Íæπ ÎàÑÎ•¥Í∏∞)
print(torch.max(A,dim=0, keepdims=True)) # Ï∞®Ïõê Ïú†ÏßÄ (Í≤∞Í≥ºÍ∞Ä [1, 4] Î™®ÏñëÏù¥ Îê®)
print(torch.max(A,dim=1, keepdims=True)) # Ï∞®Ïõê Ïú†ÏßÄ (Í≤∞Í≥ºÍ∞Ä [3, 1] Î™®ÏñëÏù¥ Îê®)
print(torch.min(A))
print(torch.min(A,dim=0)) # Í∞Å 'Ïó¥'Ïùò ÏµúÏÜüÍ∞í
print(torch.min(A,dim=1)) # Í∞Å 'Ìñâ'Ïùò ÏµúÏÜüÍ∞í
print(torch.argmax(A))

print(torch.argmax(A,dim=0)) # Í∞Å Ïó¥ÏóêÏÑú Í∞ÄÏû• ÌÅ∞ Í∞íÏùò Ïù∏Îç±Ïä§
print(torch.argmax(A,dim=1)) # Í∞Å ÌñâÏóêÏÑú Í∞ÄÏû• ÌÅ∞ Í∞íÏùò Ïù∏Îç±Ïä§
```

```python
A = torch.randint(1,5,size=(12,)) # 1Î∂ÄÌÑ∞ 5ÎØ∏Îßå 12Í∞ú Ï†ïÏàò (1Ï∞®ÏõêÏùÄ (N,)ÏúºÎ°ú ÌëúÌòÑ)
print(A)
print(A.shape)

B = A.reshape(2,2,3) # ‚òÖÏõêÏÜå Ï¥ù Í∞úÏàò(12Í∞ú)Í∞Ä ÎßûÏïÑÏïº Í∞ÄÎä•!
C = A.reshape(2,6)
print(B)
print(B.ndim) # Ï∞®ÏõêÏùò Ïàò(Rank) ÌôïÏù∏
print(C.ndim)
```

```python
a = torch.tensor([1,2,3])
b = torch.tensor([2,2,1])
print(torch.sum(a*b)) # ÎÇ¥Ï†Å (ÏÑ±Î∂ÑÎ≥Ñ Í≥± -> Ìï©)

a = a.reshape(3,1)
b = b.reshape(3,1)
# ÏïÑÎûò 4Ï§ÑÏùÄ Î™®Îëê Í∞ôÏùÄ Ïó∞ÏÇ∞ (Ï†ÑÏπò ÌõÑ ÌñâÎ†¨Í≥± = ÎÇ¥Ï†Å)
print(a.transpose(1,0)@b) # 1Î≤à Ï∂ïÍ≥º 0Î≤à Ï∂ïÏùÑ 'ÎßûÍµêÌôò' (Swap) -> Ïò§ÏßÅ 2Í∞úÎßå Í∞ÄÎä•
print(a.permute(1,0)@b) # 0Î≤à, 1Î≤à Ï∂ïÏùÑ [1, 0] ÏàúÏÑúÎ°ú 'Ïû¨Î∞∞Ïó¥' (Reorder) -> Ï†ÑÏ≤¥ ÏàúÏÑú ÏßÄÏ†ï
print(a.T@b) # Ï†ÑÏπòÌñâÎ†¨
print(a.t()@b)

A = torch.randn(4,3,6)
# Î™©Ìëú: (4, 3, 6) -> (4, 6, 3)
print(A.transpose(2,1).shape) # 2Î≤à Ï∂ïÍ≥º 1Î≤à Ï∂ïÏùò 'ÏûêÎ¶¨Î•º Î∞îÍøà' (Swap)
print(A.permute(0,2,1).shape) # Î™®Îì† Ï∂ïÏùÑ [0Î≤à, 2Î≤à, 1Î≤à] ÏàúÏÑúÎ°ú 'Ï§Ñ ÏÑ∏ÏõÄ' (Reorder)
```

```python
# Ïù∏Îç±Ïã± : Rank Í∞êÏÜå, Ïä¨ÎùºÏù¥Ïã± : Rank Ïú†ÏßÄ
x = torch.randn(2,3,4,5,6)
# 1. Îí§Ï™Ω Ï∞®Ïõê Ïãπ Îã§ Í∞ÄÏ†∏Ïò§Í∏∞
print(x[1,2,:,:,:].shape) # ÏïûÏùò Îëê Ï∞®Ïõê(0,1Î≤à)ÏùÄ ÌäπÏ†ïÌï¥ÏÑú ÏÇ¨ÎùºÏßê
print(x[1,2,...].shape)   # ...Ïù¥ ÎÇ®ÏùÄ Îí§Ï™Ω(:, :, :)ÏùÑ ÏûêÎèôÏúºÎ°ú Ï±ÑÏõåÏ§å

# 2. ÏïûÏ™Ω Ï∞®Ïõê Ïãπ Îã§ Í∞ÄÏ†∏Ïò§Í∏∞
print(x[:,:,:,:,3].shape) # ÎßàÏßÄÎßâ Ï∞®Ïõê(4Î≤à)Îßå ÌäπÏ†ïÌï¥ÏÑú ÏÇ¨ÎùºÏßê
print(x[...,3].shape)     # ...Ïù¥ ÏïûÏ™Ω(:, :, :, :)ÏùÑ ÏûêÎèôÏúºÎ°ú Ï±ÑÏõåÏ§å

# 3. Ï§ëÍ∞Ñ Ï∞®Ïõê Ïãπ Îã§ Í∞ÄÏ†∏Ïò§Í∏∞ (Ïù¥Í≤å Ï†úÏùº Ìé∏Ìï®!)
print(x[1,:,:,3,:].shape) # 0Î≤àÍ≥º 3Î≤à Ï∞®ÏõêÎßå ÌäπÏ†ïÌï¥ÏÑú ÏÇ¨ÎùºÏßê
print(x[1,...,3,:].shape) # ...Ïù¥ Ï§ëÍ∞ÑÏóê ÎÇÄ Ï∞®ÏõêÎì§(:, :)ÏùÑ ÏïåÏïÑÏÑú Ï±ÑÏõåÏ§å
```

```python
A = torch.ones(3,4)
B = torch.zeros(3,4)
# 1. ÏúÑÏïÑÎûòÎ°ú Î∂ôÏù¥Í∏∞ (Vertical, ÌñâÏù¥ ÎäòÏñ¥ÎÇ®)
C = torch.vstack([A,B]) # vstack: Vertical(ÏÑ∏Î°ú)Î°ú ÏåìÍ∏∞
E = torch.cat([A,B], dim=0) # cat(dim=0): 0Î≤à Ï∂ï(ÏÑ∏Î°ú) Î∞©Ìñ•ÏúºÎ°ú Ïó∞Í≤∞ ‚òÖ(3Ï∞®Ïõê Ïù¥ÏÉÅÏóêÏÑú Ïù¥Í±∏ Ï£ºÎ°ú ÏîÄ)
print(f"=== ÏÑ∏Î°ú Ìï©Ï≤¥ (vstack / dim=0) ===")
print(C)
print(E)
print(f"Shape: {C.shape}") # (3+3, 4) -> (6, 4)

# 2. ÏñëÏòÜÏúºÎ°ú Î∂ôÏù¥Í∏∞ (Horizontal, Ïó¥Ïù¥ ÎäòÏñ¥ÎÇ®)
D = torch.hstack([A,B]) # hstack: Horizontal(Í∞ÄÎ°ú)Î°ú ÏåìÍ∏∞
F = torch.cat([A,B], dim=1) # cat(dim=1): 1Î≤à Ï∂ï(Í∞ÄÎ°ú) Î∞©Ìñ•ÏúºÎ°ú Ïó∞Í≤∞

print(f"\n=== Í∞ÄÎ°ú Ìï©Ï≤¥ (hstack / dim=1) ===")
print(D)
print(F)
print(f"Shape: {D.shape}") # (3, 4+4) -> (3, 8)
```

```python
A = torch.randn(1,1,1,3,1,1,1,4,1,1,1) # Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞Îäî 3Í≥º 4ÎøêÏù¥Í≥† ÎÇòÎ®∏ÏßÄÎäî Îã§ Ìè¨Ïû•ÏßÄ(1)ÏûÑ
print(A.shape)
print(A.squeeze().shape) # squeeze() : Ïù∏ÏûêÍ∞Ä ÏóÜÏúºÎ©¥ 'Î™®Îì†' 1ÏùÑ Îã§ ÏóÜÏï±ÎãàÎã§. (ÏßÑÍ≥µ Ìè¨Ïû•)
print(A.squeeze(dim=(0,2,4,5)).shape) # squeeze(dim=...) : ÎÇ¥Í∞Ä 'ÏßÄÏ†ïÌïú ÏúÑÏπò'Ïùò 1Îßå ÏóÜÏï±ÎãàÎã§.

B = torch.randn(3,4)
print("======unsqueeze (ÌäπÏ†ï ÏúÑÏπòÏóê 1 ÎÅºÏõåÎÑ£Í∏∞)======")
print(B.unsqueeze(dim=0).shape) # Îß® Ïïû Ï∂îÍ∞Ä -> (1, 3, 4) ‚òÖ Ï†úÏùº ÎßéÏù¥ ÏîÄ (Î∞∞Ïπò Ï∂îÍ∞Ä)
print(B.unsqueeze(dim=1).shape) # Ï§ëÍ∞Ñ Ï∂îÍ∞Ä -> (3, 1, 4)
print(B.unsqueeze(dim=2).shape) # Îß® Îí§ Ï∂îÍ∞Ä -> (3, 4, 1)

print("======reshape (ÏàòÎèôÏúºÎ°ú Î™®Ïñë Ïû°Í∏∞)======")
# unsqueezeÎûë Í≤∞Í≥ºÎäî ÎòëÍ∞ôÏùå. Îã®ÏßÄ Î™®Îì† Ïà´ÏûêÎ•º ÎÇ¥Í∞Ä Îã§ Ï†ÅÏñ¥Ïïº Ìï®.
print(B.reshape(1,3,4).shape) # (1, 3, 4)
print(B.reshape(3,1,4).shape) # (3, 1, 4)
print(B.reshape(3,4,1).shape) # (3, 4, 1)
```

```python
A = torch.ones(3,4)
B = torch.zeros(3,4)
C = torch.stack([A,B]) # stack: "ÏÉàÎ°úÏö¥ Ï∞®ÏõêÏùÑ ÎßåÎì§Ïñ¥ÏÑú Ìè¨Í∞úÍ∏∞" (ÎëêÍ∫ºÏõåÏßê) -> Ï∞®Ïõê Ï¶ùÍ∞Ä (+1)
D = torch.cat([A,B], dim=0) # cat (Concatenate): "Í∏∞Ï°¥ Ï∞®ÏõêÏóêÏÑú Ïù¥Ïñ¥ Î∂ôÏù¥Í∏∞" (ÎÑìÏñ¥ÏßÄÍ±∞ÎÇò Í∏∏Ïñ¥Ïßê) -> Ï∞®Ïõê Ïú†ÏßÄ
print(C)
print(C.shape)
print(D)
print(D.shape)
```
